NumPy Coding Tasks

1.Create Arrays

```python
import numpy as np
arr=np.array([1,2,3,4])
print(arr)
```

2.Array Operations

```python
a=np.array([1,2,3])
b=np.array([3,4,5])
print(a+b)
print(a*b)
```

3.Zeros and Ones

```python
print(np.zeros((2,3)))
print(np.ones((3,3)))
```

4.Mean, Max, Min

```python
arr=np.array([1,2,3,4,5])
print(arr.mean())
print(arr.max())
print(arr.min())
```
