**Spotify Tracks Dataset**

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("spotify_dataset.csv")
df.shape
```

```python
df=df.sample(40000,random_state=42)
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
df.sample(3)
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df.to_csv("spotify_dataset_Cleaned.csv", index=False)
```

```python
#Univariate Analysis
import matplotlib.pyplot as plt

df["track_genre"].value_counts().head(15).plot(kind="bar",figsize=(10,5))
plt.title("Top 15 Music Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Tracks")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["popularity"], bins=20)
plt.title("Popularity Distribution")
plt.xlabel("Popularity")
plt.ylabel("Frequency")
plt.show()
```

```python
#Outlier Detection
import seaborn as sns

sns.boxplot(y=df["tempo"])
plt.title("Tempo Boxplot")
plt.ylabel("Tempo")
plt.show()
```

```python
#Bivariate Analysis
df.groupby("track_genre")["popularity"].mean().sort_values(ascending=False).head(15).plot(kind="bar", figsize=(10,5))
plt.title("Top 15 Genres by Average Popularity")
plt.xlabel("Genre")
plt.ylabel("Average Popularity")
plt.show()
```

```python
#Correlation Analysis
numeric_df=df.select_dtypes(include=["int64","float64"])
corr=numeric_df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
