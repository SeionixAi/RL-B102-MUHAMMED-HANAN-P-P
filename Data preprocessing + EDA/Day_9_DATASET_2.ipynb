```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("hotel_bookings.csv")
df=df.sample(10000,random_state=10)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
print(df.duplicated().sum())
```

```python
df.isnull().sum()
```

```python
df['children'] = df['children'].fillna(df['children'].median())
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)
```

```python
df.sample(4)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
print(df['is_canceled'].mean())

cancel_by_hotel = df.groupby('hotel')['is_canceled'].mean()
cancel_by_hotel.plot(kind='bar', color=['#4C72B0','#DD8452'])
plt.title("Cancellation Rate by Hotel Type")
plt.ylabel("Cancellation Rate")
plt.show()
```

```python
month_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=month_order, ordered=True)

monthly_cancel = df.groupby('arrival_date_month')['is_canceled'].mean()
monthly_cancel.plot(kind='line', marker='o')
plt.title("Cancellation Rate by Month")
plt.ylabel("Cancellation Rate")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='is_canceled', y='lead_time', data=df)
plt.title("Lead Time vs Cancellation")
plt.xlabel("Canceled (0=No, 1=Yes)")
plt.ylabel("Lead Time (days)")
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar')
plt.title("Top 10 Countries by Bookings")
plt.ylabel("Number of Bookings")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='hotel', y='adr', data=df[df['adr'] < 500])
plt.title("ADR by Hotel Type")
plt.show()
```

```python
num_cols = ['lead_time','stays_in_weekend_nights','stays_in_week_nights','adults','children','previous_cancellations','adr','is_canceled']
plt.figure(figsize=(8,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
