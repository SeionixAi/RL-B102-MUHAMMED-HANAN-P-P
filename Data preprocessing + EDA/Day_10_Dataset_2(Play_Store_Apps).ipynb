**Google Play Store Apps**

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Google-Playstore.csv")
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.sample(30000,random_state=42)
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
df.fillna("Unknown",inplace=True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.sample(3)
```

```python
df.to_csv("Google-Playstore_cleaned.csv",index=False)
```

```python
#Univariate Analysis
import matplotlib.pyplot as plt

df["Category"].value_counts().plot(kind="bar", figsize=(12,5))
plt.title("App Category Distribution")
plt.xlabel("Category")
plt.ylabel("Number of Apps")
plt.show()
```

```python
plt.hist(df["Price"], bins=3)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Rating"], bins=5)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
df["Free"].value_counts().plot(kind="bar")
plt.title("Free vs Paid Apps")
plt.xlabel("Free")
plt.ylabel("Number of Apps")
plt.show()
```

```python
#Outlier Detection
import seaborn as sns

sns.boxplot(y=df["Price"])
plt.title("Price Boxplot")
plt.ylabel("Price")
plt.show()
```

```python
#Bivariate Analysis
plt.figure(figsize=(8,6))
plt.scatter(df["Rating Count"],df["Rating"])
plt.title("Rating Count vs Rating")
plt.ylabel("Rating")
plt.xlabel("Rating Count")
plt.show()
```

```python
df.groupby("Category")["Rating"].mean().sort_values().plot(kind="bar",figsize=(12,5
                                                                 ))
plt.title("Average Rating by Category")
plt.ylabel("Average Rating")
plt.xlabel("Category")
plt.show()
```

```python
df.groupby("Category")["Price"].mean().sort_values().plot(kind="bar",figsize=(12,5))
plt.title("Average Price by Category")
plt.xlabel("Category")
plt.ylabel("Average Price")
plt.show()
```

```python
#Correlation Analysis
import seaborn as sns
numeric_df=df.select_dtypes(include=["int64","float64"])
corr=numeric_df.corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
