```python
import matplotlib.pyplot as plt
```

```python
#Univariate Analysis
df["room_type"].value_counts().plot(kind="bar")

plt.xlabel("Room Type")
plt.ylabel("Number of lisitngs")
plt.title("Room Type Distribution")
plt.show()
```

```python
df["neighbourhood_group"].value_counts().plot(kind="bar")

plt.xlabel("Neighbourhood Group")
plt.ylabel("Number of lisitngs")
plt.title("Neighbourhood Group Distribution")
plt.show()
```

```python
plt.hist(df["price"],bins=30)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()
```

```python
plt.hist(df["availability_365"],bins=30)
plt.xlabel("Available Days")
plt.ylabel("Frequency")
plt.title("Availability Distribution")
plt.show()
```

```python
#Outlier Detection
import seaborn as sns
sns.boxplot(y=df["price"])
plt.ylabel("Price")
plt.title("Price Boxplot")
plt.show()
```

```python
#Bivariate Analysis
plt.scatter(df["number_of_reviews"],df["price"])
plt.xlabel("Number of Reviews")
plt.ylabel("Price")
plt.title("Number of Reviews vs Price")
plt.show()
```

```python
df.groupby("room_type")["price"].mean().plot(kind="bar")
plt.xlabel("Room Type")
plt.ylabel("Average Price")
plt.title("Average Price by Room Type")
plt.show()
```

```python
df.groupby("neighbourhood_group")["price"].mean().plot(kind="bar")

plt.title("Average Price by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price")

plt.show()
```

```python
#Correlation Analysis
numeric_df=df.select_dtypes(include=["int64","float64"])
corr=numeric_df.corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
