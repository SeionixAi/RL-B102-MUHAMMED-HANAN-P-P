```python
df['Order Date']=pd.to_datetime(df['Order Date'],dayfirst=True)
df['Ship Date']=pd.to_datetime(df['Ship Date'],dayfirst=True)
df.sample(2)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
#Category Distribution
df['Category'].value_counts().plot(kind="bar")
plt.xlabel("Category")
plt.ylabel("Number of orders")
plt.title("Category Distribution")
plt.show()
```

```python
#Sales Distribution
plt.hist(df['Sales'],bins=20)
plt.xlabel("Sales amount")
plt.ylabel("Frequency")
plt.title("Sales Distribution")
plt.show()
```

```python
#Profit Distribution
plt.hist(df['Profit'],bins=10)
plt.xlabel("Profit amount")
plt.ylabel("Frequency")
plt.title("Profit Distribution")
plt.show()
```

```python
# Sales vs Profit
plt.scatter(df['Sales'],df['Profit'])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
# Average Profit by Category
df.groupby("Category")["Profit"].mean().plot(kind="bar")
plt.title("Average Profit by Category")
plt.xlabel("Category")
plt.ylabel("Average Profit")
plt.show()
```

```python
# Correlation Heatmap
import seaborn as sns
numeric_df=df[["Sales","Quantity", "Discount", "Profit", "Shipping Cost"]]
corr=numeric_df.corr()
sns.heatmap(corr,annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Sales Boxplot
sns.boxplot(x=df["Category"],y=df["Sales"])
plt.title("Sales Boxplot")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
# Top 10 Products by Sales
top_products=df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
top_products.plot(kind="bar")
plt.title("Top 10 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Sales")
plt.show()
```

```python
# Region-wise Sales
region_sales=df.groupby("Region")["Sales"].sum().sort_values().plot(kind="bar")
plt.title("Region-wise Sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```
