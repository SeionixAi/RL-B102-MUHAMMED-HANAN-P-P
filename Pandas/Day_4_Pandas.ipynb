```python
# 5.Handling Missing Data
df.isnull().sum() # Count nulls
```

```python
df.dropna(inplace=True) # Drop rows with any nulls
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True) # Fill with mean
```

```python
# 6.Removing Duplicates
df.duplicated().sum()
df.dropna(inplace=True)
```

```python
# 7.Selecting Columns
df['Age'] # Single column (Series)
```

```python
df[['Name','Age']] # Multiple columns (DataFrame)
```

```python
# 8.Filtering Rows
df[df['Age'] > 30] # Age > 30
```

```python
df[df['Sex'] == 'female'] # All females
```

```python
df[(df['Age'] > 30) & (df['Sex'] == 'male')] # Compound filter
```

```python
# 9.Sorting Data
df.sort_values(by='Age', ascending=True)
```

```python
df.sort_values(by=['Pclass', 'Fare'], ascending=[True, False])
```

```python
# 10.Creating New Columns
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df.sample(2)
```

```python
# 11.Dropping Columns or Rows
df.drop('Ticket', axis=1, inplace=True) # Drop column
df.drop([0, 1], axis=0, inplace=True) # Drop rows by index
```

```python
df.head()
```

```python
# 12.Renaming Columns
df.rename(columns={'SibSp': 'SiblingsSpouses'}, inplace=True)
```

```python
df.sample(2)
```

```python
# 13.Value Counts & Unique
df['Sex'].value_counts()
```

```python
df['Embarked'].unique()
```

```python
df['Embarked'].nunique()
```

```python
# 14.Grouping and Aggregation
df.groupby('Pclass')['Fare'].mean()
```

```python
df.groupby('Sex')['Survived'].sum()
```

```python
df.groupby(['Pclass', 'Sex'])['Age'].median()
```

```python
# 15. Aggregating Multiple Columns
df.groupby('Pclass').agg({
'Age': ['mean', 'min'],
'Fare': ['max', 'mean']
})
```

```python
# 16. Apply Custom Functions
df['AgeSquared'] = df['Age'].apply(lambda x: x**2)
```

```python
df.sample(1)
```

```python
# 17. String Operations
df['Name'].str.lower()
```

```python
df['Name'].str.contains('Mrs')
```

```python
# 18. Mapping & Replacing Values
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
df['Embarked'].replace({'S': 'Southampton', 'C': 'Cherbourg'}, inplace=True)
```

```python
df.sample(2)
```

```python
# 19. Combining Columns (concatenation)
df['FullName'] = df['Name'] + ' (' + df['Sex'].astype(str) + ')'
```

```python
df.sample(2)
```

```python
# 20. Merging DataFrames
import numpy as np
df12 = pd.DataFrame({
    "PassengerId": np.arange(1, 201),
    "City": np.random.choice(["Delhi", "Mumbai", "Chennai", "Kolkata"], 200),
    "Salary": np.random.randint(30000, 100000, 200)
})

df12.to_csv("additional.csv", index=False)
df2 = pd.read_csv('additional.csv')
df = pd.merge(df, df2, on='PassengerId', how='left')
```

```python
df.head()
```

```python
# 21. Concatenating DataFrames
df_combined = pd.concat([df, df2], axis=0) # Stack vertically
print(df_combined)
```

```python
# 22. Pivot Table
pd.pivot_table(df, values='Fare', index='Pclass', columns='Sex', aggfunc='mean')
```

```python
# 23. Resetting and Setting Index
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df.set_index('PassengerId', inplace=True)
```

```python
df.head()
```

```python
# 24. Conditional Column Creation
df['AgeGroup'] = df['Age'].apply(lambda x: 'child' if x < 16 else 'adult')
```

```python
df.head()
```

```python
# 25. Save Cleaned Data
df.to_csv('cleaned_titanic.csv', index=False)
df.to_excel('cleaned_data.xlsx', index=False)
```
