break, continue, and pass

```python
for i in range(1,11):
  if i==3:
    continue
  if i==5:
    break
  if i==7:
    pass
  print(i)
```

Recursion

```python
def fact(n):
  if n==0:
    return 1
  return n*fact(n-1)
fact(5)
```

Lambda Function

```python
sqr=lambda a:a*a
sqr(4)
```

is vs ==

```python
a=[1,2,3]
b=[1,2,3]
print(a==b)
print(a is b)
```
