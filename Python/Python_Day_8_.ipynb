1. Find the Factorial of a Number

```python
def fact1(n): #Recursive
  if n==0:
    return 1
  else:
    return n*fact(n-1)

def fact2(n): #Iterative
  result=1
  for i in range(1,n+1):
    result*=i
  return result

print(fact(6))
print(fact1(0))
```

2. Find the Largest and Smallest Number in a List

```python
def func(s):
  smallest=s[0]
  largest=s[0]
  for num in s:
    if num > largest:
      largest=num
    if num < smallest:
      smallest=num
  return smallest,largest

print(func([1,2,3,4,0]))
```

3. Fibonacci Sequence up to N terms

```python
def fib(n):
  a,b=0,1
  for i in range(n):
    print(a,end=" ")
    x=b
    b=a+b
    a=x
fib(10)
```

4. Create a Square Frame Pattern

```python
def sq_frame(n):
  for i in range(n):
    for j in range(n):
      if i==0 or i==n-1 or j==0 or j==n-1:
        print("*",end="")
      else:
        print(" ",end="")
    print()
sq_frame(5)
```

5. Sum of Digits in a Number

```python
def sum_of_digits(s):
  sum=0
  for num in s:
    sum+=int(num)
  return sum
print(sum_of_digits("123"))
```
