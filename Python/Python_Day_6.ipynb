1. Write a Python program to count the number of consonants in a given string.


```python
def count_consonants(str):
  vowels = ['a', 'e', 'i', 'o', 'u']
  count=0
  for x in str.lower():
    if x not in vowels:
      if x!=' ':
        count+=1
  return count
print(count_consonants("Muhammed Hanan"))
```

2. Write a Python program to find the first non-repeating character in a string.


```python
def non_rpt(str):
  char_freq={}
  for char in str.lower():
    if char in char_freq:
      char_freq[char]+=1
    else:
      char_freq[char]=1
  for char in str.lower():
    if char_freq[char]==1:
      return char
  return None
print(non_rpt("Muhammed Hanan"))
```

3.Write a Python program to remove all duplicate characters from a string.

```python
def rem_dup(str):
  result=""
  for char in str.lower():
    if char not in result:
      result+=char
  return result
print(rem_dup("Muhammed Hanan"))
```

4. Write a Python program to count the frequency of each character in a string.


```python
def freq_count(str):
  freq={}
  for char in str:
    if char in freq:
      freq[char]+=1
    else:
      freq[char]=1
  return freq
print(freq_count("Muhammed Hanan"))
```
