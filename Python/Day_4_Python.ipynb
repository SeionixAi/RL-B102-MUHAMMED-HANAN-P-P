1.Write a Python program to count the number of vowels in a given string.

```python
def count_vowels(text):
  vowels="aeiou"
  count=0
  for char in text.lower():
    if char in vowels:
      count+=1
  return count

print(count_vowels("Muhammed Hanan P P"))
```

2.Write a Python program to check whether a given string is a palindrome.

```python
def is_palindrome(text):
  lower_text=text.lower()
  rev=lower_text[::-1]
  if(lower_text==rev):
    print(text," is a palindrome")
  else:
    print(text," is not a palindrome")
is_palindrome("malayalam")
is_palindrome("Hello")
```

3.Write a Python program to check whether two strings are anagrams.

```python
def is_anagram(str1,str2):
  l1=str1.lower()
  l2=str2.lower()
  sorted_str1=sorted(l1)
  sorted_str2=sorted(l2)
  if(sorted_str1==sorted_str2):
    print(str1," and ",str2," are anagrams")
  else:
    print(str1," and ",str2," are not anagrams")
is_anagram("hanan","nahan")
is_anagram("listen","Silent")
is_anagram("hello","world")
```
