1. Write a Python program to check whether a string contains only digits.

```python
def is_only_digits(str):
  digit="0123456789"
  for char in str:
    if char not in digit:
      return False
  return True
print(is_only_digits("123456789"))
print(is_only_digits("123456789a"))
print(is_only_digits("Muhammed Hanan"))
```

2. Write a Python program to find the longest word in a string.

```python
def longest(str):
  size_word={}
  highest=0
  highest_word=""
  splitted_str=str.split(" ")
  for word in splitted_str:
    if word not in size_word:
      size_word[word]=len(word)
      if len(word)>highest:
        highest=len(word)
        highest_word=word
  print(size_word)
  return print("Highest word is ",highest_word," with length ",highest)
print(longest("Muhammed Hanan P P is my name"))

```

3. Write a Python program to capitalize the first and last character of each word in a string.

```python
def capt(str):
  words=str.split(" ")
  result=""
  for word in words:
    if(len(word)>1):
      result+=word[0].upper()+word[1:-1]+word[-1].upper()+" "
    else:
      result+=word.upper()+" "
  return result
print(capt("my name is muhammed hanan p p"))
```

4. Draw a right-angled triangle pattern using stars.

```python
def right_triangle(n):
  for i in range(1,n+1):
    print("*"*i)
print(right_triangle(3))
```

5. Draw an inverted right-angled triangle pattern.

```python
def inv_right_triangle(n):
  for i in range(n,0,-1):
    print("*"*i)
print(inv_right_triangle(3))
```
